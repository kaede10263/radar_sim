# === 1. Generate ADC data ===

```python
def generate_adc_data_cube(Q, distance, velocity, Theta, Phi):
    c = 3e+8                                # 光速
    fc = 77e+9                              # 發射訊號載頻
    Tc = 10e-6                              # 時間週期
    fs = 25.6e+6                            # 取樣頻率(sample frequence)/取樣率(sampling rate)
    Fs = 1/fs                               # 取樣週期(sampling period)/取樣間隔(sampling interval)
    t = np.arange(0, Tc, Fs)                # 時間向量 (共256個點
    Sample_rate = len(t)                    # 取樣頻率256個點
    bw = 500e+6                             # 發射訊號帶寬
    k = bw/Tc                               # 斜率
    P = 100                                 # chirp 的數目
    # distance = np.array([10, 30, 50])       # 目標距離雷達50m
    # velocity = np.array([-20, 0, 20])       # 目標距離雷達的相對速度20m/s
    # Theta = np.array([-50, 0, 50])          # 目標入射角度
    # Phi = np.array([-30, -20, -10])         # 目標發射角度
    fb = 2*k*distance/c                     # 訊號關於距離頻率
    fd = 2*velocity*fc/c                    # 訊號關於速度頻率
    M = 4                                   # 接收天線數目
    N = 3                                   # 發射天線數目
    # dm = np.arange(0, M, 1);                # 接收天線的索引位置
    # dn = np.arange(0, N, 1);                # 發射天線的索引位置
    lamda = c/fc                            # 訊號波長
    d = lamda/2                             # 天線間距
    # Q = 3;                                  # 目標數目
    SNR = 20;                               # SNR
    Np = 1                                  # 雜訊功率
    An = np.sqrt(Np)                        # 雜訊振幅
    Sp = Np*10**(SNR/10)                    # 訊號功率
    As = np.sqrt(Sp)                        # 訊號振幅
    #------------------------------- 產生資料 ---------------------------------#
    Element = 0;                            # 天線數目MxN
    adc_data_cube = np.zeros((Sample_rate, P, (M*N)), dtype=np.complex64)                # 存儲所有天線的中頻訊號
    for n in range(N):                                                                   # 發射天線迴圈
        for m in range(M):                                                               # 接收天線迴圈
            data = np.zeros((Sample_rate, P), dtype=np.complex64)                        # 存儲所有目標的中頻訊號
            for q in range(Q):
                IF_data = np.zeros((Sample_rate, P), dtype=np.complex64)                 # 存儲含有雜訊的中頻訊號
                for p in range(P):
                    timeSignal = np.exp(1j*2*np.pi*((fb[q]+fd[q])*t))                    # 含時間的訊號
                    velocitySignal = np.exp(1j*2*np.pi*fd[q]*(p+1)*Tc)                   # 含速度的訊號
                    distanceSignal = np.exp(1j*2*np.pi*fc*2*distance[q]/c)               # 含距離的訊號
                    DOD = np.exp(1j*2*np.pi*n*d*np.sin((Phi[q]*np.pi/180))/lamda)        # 離開角度訊號
                    DOA = np.exp(1j*2*np.pi*m*d*np.sin((Theta[q]*np.pi/180))/lamda)      # 入射角度訊號
                    IF = As*(timeSignal*velocitySignal*distanceSignal*DOD*DOA);
                    # noise = An*(np.random.randn(1, Sample_rate)+1j*np.random.randn(1, Sample_rate))/np.sqrt(2)
                    # IF_data[:, p] = IF+noise;
                    IF_data[:, p] = IF;
                #-----------------------------------------------------------------#
                data = data+IF_data                                                      # 訊號疊加
            #-----------------------------------------------------------------#
            adc_data_cube[:, :, Element] = data
            Element = Element+1
    
    save_dict = {
    "adc_data_cube": adc_data_cube,
    # "theta": theta
    }
    save_path = "adc_data_with_theta.npy"
    np.save(save_path, save_dict)
    return adc_data_cube
```

# === 2. Range and doppler FFT  ===

```python
def RV_fft(adc_data_cube, plot=True):
    # 輸出 shape: 256 range bins, 100 chirps, 12 antenna (3x4)
    c = 3e+8                                # 光速
    fc = 77e+9                              # 發射訊號載頻
    Tc = 10e-6                              # 時間週期
    fs = 25.6e+6                            # 取樣頻率(sample frequence)/取樣率(sampling rate)
    Fs = 1/fs                               # 取樣週期(sampling period)/取樣間隔(sampling interval)
    t = np.arange(0, Tc, Fs)                # 時間向量 (共256個點)
    Sample_rate = len(t)                    # 取樣頻率256個點
    bw = 500e+6                             # 發射訊號帶寬
    k = bw/Tc                               # 斜率
    P = 100                                 # chirp 的數目
    M = 4                                   # 接收天線數目
    N = 3                                   # 發射天線數目
    lamda = c/fc                            # 訊號波長
    # distance_temp = ((0:Sample_rate-1)*fs*c)/(2*k*Sample_rate);
    #----------------------------------------------------------------------------------------------------#
    dataSum = np.zeros((Sample_rate, P), dtype=np.complex64)
    for i in range(M*N):
        dataSum = dataSum+adc_data_cube[:, :, i]
        dataAvg = dataSum/(M*N)                                                  # 所有天線接收資料的總和，資料維度256x100
    
    
    range_spectrum = np.fft.fft(dataAvg, axis=0)                             # 估測的距離頻譜
    doppler_spectrum = np.fft.fft(range_spectrum, axis=1)            # 估測的速度頻譜
    rvSpectrum_shift = np.fft.fftshift(doppler_spectrum, axes=1)# 將頻域中心化（方便觀察）
    rvSpectrum = np.abs(rvSpectrum_shift)
    
    range_res = c * fs / (2 * k * Sample_rate)         # 每個 range bin 的距離解析度
    max_range = range_res * Sample_rate                # 最遠距離
    ranges = np.linspace(0, max_range, Sample_rate)      # y 軸：range bin 對應的距離
    
    # Doppler 計算
    doppler_res = 1 / (P * Tc)                         # 頻率解析度
    max_doppler = doppler_res * P / 2
    velocities = np.linspace(-max_doppler, max_doppler, P) * lamda / 2  # x 軸：Doppler bins 對應速度
    
    # 生成 meshgrid 對應 range 和 velocity 軸
    V, R = np.meshgrid(velocities, ranges)
    Z = rvSpectrum
    
    # 畫 3D 圖
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(V, R, Z, cmap='viridis')
    
    ax.set_xlabel('Velocity (m/s)')
    ax.set_ylabel('Range (m)')
    ax.set_zlabel('Magnitude (dB)')
    ax.set_title('3D Range-Doppler Map')
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Magnitude (dB)')
    ax.view_init(elev=90, azim=0)
    plt.tight_layout()
    plt.show()
    
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(V, R, Z, cmap='viridis')
    
    ax.set_xlabel('Velocity (m/s)')
    ax.set_ylabel('Range (m)')
    ax.set_zlabel('Magnitude (dB)')
    ax.set_title('3D Range-Doppler Map')
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Magnitude (dB)')
    ax.view_init(elev=0, azim=0)
    plt.tight_layout()
    plt.show()
    return rvSpectrum
```

# === 3. CFAR  ===

```python
def cfar_detect(rvSpectrum, guard_cells=2, training_cells=8, threshold_scale=15):
    c = 3e+8                                # 光速
    fc = 77e+9                              # 發射訊號載頻
    Tc = 10e-6                              # 時間週期
    fs = 25.6e+6                            # 取樣頻率(sample frequence)/取樣率(sampling rate)
    Fs = 1/fs                               # 取樣週期(sampling period)/取樣間隔(sampling interval)
    t = np.arange(0, Tc, Fs)                # 時間向量 (共256個點)
    Sample_rate = len(t)                    # 取樣頻率256個點
    bw = 500e+6                             # 發射訊號帶寬
    k = bw/Tc                               # 斜率
    P = 100                                 # chirp 的數目
    M = 4                                   # 接收天線數目
    N = 3                                   # 發射天線數目
    lamda = c/fc                            # 訊號波長
    # distance_temp = ((0:Sample_rate-1)*fs*c)/(2*k*Sample_rate);
    #----------------------------------------------------------------------------------------------------#
    
    detections = []
    num_range_bins, num_doppler_bins = rvSpectrum.shape
    T = training_cells
    G = guard_cells

    for r in range(T + G, num_range_bins - (T + G)):
        for d in range(1, num_doppler_bins - 1):
            training_left = rvSpectrum[r - T - G:r - G, d]
            training_right = rvSpectrum[r + G + 1:r + G + 1 + T, d]
            noise_level = np.mean(np.concatenate([training_left, training_right]))
            threshold = threshold_scale * noise_level
            cell_value = rvSpectrum[r, d]
            is_local_peak = (cell_value > rvSpectrum[r, d - 1]) and (cell_value > rvSpectrum[r, d + 1])
            if cell_value > threshold and is_local_peak:
                detections.append({
                    "range_bin": r,
                    "doppler_bin": d,
                    "magnitude": cell_value
                })

    # 建立對應的 range 與 velocity 軸
    range_res = c * fs / (2 * k * Sample_rate)
    doppler_res = 1 / (P * Tc)
    ranges = np.linspace(0, range_res * Sample_rate, Sample_rate)
    velocities = np.linspace(-P / 2, P / 2 - 1, P) * doppler_res * lamda / 2
    V, R = np.meshgrid(velocities, ranges)
    Z = rvSpectrum

    # 準備偵測點資料（轉換為實際單位）
    detect_ranges = [ranges[d["range_bin"]] for d in detections]
    detect_velocities = [velocities[d["doppler_bin"]] for d in detections]
    detect_mags = [d["magnitude"] for d in detections]

    # 繪製 3D 圖 + CFAR 點
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(V, R, Z, cmap='viridis', alpha=0.8)
    ax.scatter(detect_velocities, detect_ranges, detect_mags, c='r', s=40, label='CFAR detections')

    ax.set_xlabel('Velocity (m/s)')
    ax.set_ylabel('Range (m)')
    ax.set_zlabel('Magnitude')
    ax.set_title('3D Range-Doppler Map with CFAR Detections')
    ax.view_init(elev=30, azim=135)
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Magnitude')
    ax.legend()
    plt.tight_layout()
    plt.show()

    # 額外繪製：CFAR 門檻 vs Signal（擷取中心 Doppler bin）
    target_doppler_bin = P // 2
    signal_line = rvSpectrum[:, target_doppler_bin]
    thresholds = np.zeros_like(signal_line)
    for r in range(T + G, Sample_rate - (T + G)):
        training_left = signal_line[r - T - G:r - G]
        training_right = signal_line[r + G + 1:r + G + 1 + T]
        noise_level = np.mean(np.concatenate([training_left, training_right]))
        thresholds[r] = threshold_scale * noise_level

    plt.figure(figsize=(10, 4))
    plt.plot(signal_line, label='Signal', linewidth=1)
    plt.plot(thresholds, label='CFAR Threshold', color='red', linewidth=2)
    plt.xlabel('Range Bin Index')
    plt.ylabel('Magnitude')
    plt.title(f'CFAR Threshold vs Signal (Doppler Bin {target_doppler_bin})')
    plt.grid(True)
    # plt.ylim(0, 20000)
    plt.legend()
    plt.tight_layout()
    plt.show()


    plt.figure(figsize=(10, 6))
    plt.pcolormesh(velocities, ranges, rvSpectrum, shading='auto', cmap='viridis')
    plt.colorbar(label='Magnitude')
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Range (m)')
    plt.title('2D Range-Doppler Map with CFAR Detections')

    # 疊加偵測點
    plt.scatter(detect_velocities, detect_ranges, c='r', marker='x', label='CFAR Detections')
    plt.legend()
    plt.tight_layout()
    plt.show()



    return detections
```

```python
import numpy as np
import matplotlib.pyplot as plt

def cfar_detect_method1(rvSpectrum,
           guard_cells=(4, 2),
           training_cells=(8, 4),
           offset_dB=1.4,
           c=3e8, fs=25.6e6, bw=500e6, P=100, Tc=10e-6, fc=77e9):
    
    c = 3e+8                                # 光速
    fc = 77e+9                              # 發射訊號載頻
    Tc = 10e-6                              # 時間週期
    fs = 25.6e+6                            # 取樣頻率(sample frequence)/取樣率(sampling rate)
    Fs = 1/fs                               # 取樣週期(sampling period)/取樣間隔(sampling interval)
    t = np.arange(0, Tc, Fs)                # 時間向量 (共256個點)
    Sample_rate = len(t)                    # 取樣頻率256個點
    bw = 500e+6                             # 發射訊號帶寬
    k = bw/Tc                               # 斜率
    P = 100                                 # chirp 的數目
    M = 4                                   # 接收天線數目
    N = 3                                   # 發射天線數目
    lamda = c/fc                            # 訊號波長
    # distance_temp = ((0:Sample_rate-1)*fs*c)/(2*k*Sample_rate);
    #----------------------------------------------------------------------------------------------------#
    Gr, Gd = guard_cells
    Tr, Td = training_cells
    Nr, Nd = rvSpectrum.shape
    k = bw / Tc
    lamda = c / fc
    Sample_rate = Nr

    # rvSpectrum_dB = 20 * np.log10(rvSpectrum + 1e-6)
    rvSpectrum_dB = rvSpectrum / np.max(rvSpectrum)
    detection_list = []

    for i in range(Tr + Gr, Nr // 2 - (Tr + Gr)):
        for j in range(Td + Gd, Nd - (Td + Gd)):
            noise_sum = 0.0
            count = 0
            for p in range(i - (Tr + Gr), i + (Tr + Gr) + 1):
                for q in range(j - (Td + Gd), j + (Td + Gd) + 1):
                    if abs(i - p) > Gr or abs(j - q) > Gd:
                        noise_sum += rvSpectrum_dB[p, q]
                        count += 1
            if count == 0:
                continue
            noise_avg_linear = noise_sum / count
            # threshold_dB = 10 * np.log10(noise_avg_linear) + offset_dB
            threshold_dB = noise_avg_linear
            CUT_dB = rvSpectrum_dB[i, j]
            
            # if count % 100 == 0:
            print(f"db : {CUT_dB} , thres : {threshold_dB}")
                
            if CUT_dB > threshold_dB:
                detection_list.append({
                    "range_bin": i,
                    "doppler_bin": j,
                    "magnitude": rvSpectrum[i, j]
                })
            

    # 軸轉換
    range_res = c * fs / (2 * k * Sample_rate)
    ranges = np.linspace(0, range_res * Sample_rate, Sample_rate)
    doppler_res = 1 / (P * Tc)
    velocities = np.linspace(-P / 2, P / 2 - 1, P) * doppler_res * lamda / 2
    V, R = np.meshgrid(velocities, ranges)
    Z = rvSpectrum

    # 偵測點座標
    detect_ranges = [ranges[d["range_bin"]] for d in detection_list]
    detect_velocities = [velocities[d["doppler_bin"]] for d in detection_list]
    detect_mags = [d["magnitude"] for d in detection_list]

    # 🔷 1. 3D 圖 + CFAR 點
    fig = plt.figure(figsize=(12, 8))
    ax = fig.add_subplot(111, projection='3d')
    surf = ax.plot_surface(V, R, Z, cmap='viridis', alpha=0.8)
    ax.scatter(detect_velocities, detect_ranges, detect_mags, c='r', s=40, label='CFAR detections')
    ax.set_xlabel('Velocity (m/s)')
    ax.set_ylabel('Range (m)')
    ax.set_zlabel('Magnitude')
    ax.set_title('3D Range-Doppler Map with CFAR Detections')
    ax.view_init(elev=30, azim=135)
    fig.colorbar(surf, shrink=0.5, aspect=10, label='Magnitude')
    ax.legend()
    plt.tight_layout()
    plt.show()

    # 🔷 2. CFAR Threshold vs Signal
    target_doppler_bin = P // 2
    signal_line = rvSpectrum[:, target_doppler_bin]
    thresholds = np.zeros_like(signal_line)
    T = Tr
    G = Gr
    for r in range(T + G, Sample_rate - (T + G)):
        training_left = signal_line[r - T - G:r - G]
        training_right = signal_line[r + G + 1:r + G + 1 + T]
        noise_level = np.mean(np.concatenate([training_left, training_right]))
        thresholds[r] = 10 * np.log10(noise_level + 1e-6) + offset_dB

    plt.figure(figsize=(10, 4))
    plt.plot(20 * np.log10(signal_line + 1e-6), label='Signal (dB)', linewidth=1)
    plt.plot(thresholds, label='CFAR Threshold (dB)', color='red', linewidth=2)
    plt.xlabel('Range Bin Index')
    plt.ylabel('Magnitude (dB)')
    plt.title(f'CFAR Threshold vs Signal (Doppler Bin {target_doppler_bin})')
    plt.grid(True)
    plt.ylim(0, 50)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # 🔷 3. 2D Range-Doppler Map + CFAR 點
    plt.figure(figsize=(10, 6))
    plt.pcolormesh(velocities, ranges, rvSpectrum, shading='auto', cmap='viridis')
    plt.colorbar(label='Magnitude')
    plt.xlabel('Velocity (m/s)')
    plt.ylabel('Range (m)')
    plt.title('2D Range-Doppler Map with CFAR Detections')
    plt.scatter(detect_velocities, detect_ranges, c='r', marker='x', label='CFAR Detections')
    plt.legend()
    plt.tight_layout()
    plt.show()

    return detection_list
```

# === 4. AoA FFT  ===

```python
# 實作 AoA2D DPU 功能：根據 range_doppler_detection_list 從 adc_data_cube 中取出對應虛擬天線陣列資料並做 2D FFT 求 AoA

def aoa2d_dpu(adc_data_cube, detection_list, N=3, M=4):
    """
    根據 CFAR 偵測結果執行 AoA2D FFT，取得目標的 Azimuth/Elevation FFT 峰值與 Cartesian 坐標。
    Args:
        adc_data_cube: shape = (range_bin, chirp, antenna)
        detection_list: list of dicts with keys 'range_bin', 'doppler_bin'
        N: 發射天線數量
        M: 接收天線數量
    Returns:
        List of point clouds with Cartesian (x, y, z) and azimuth/elevation bins
    """
    
    c = 3e+8                                # 光速
    fc = 77e+9                              # 發射訊號載頻
    Tc = 10e-6                              # 時間週期
    fs = 25.6e+6                            # 取樣頻率(sample frequence)/取樣率(sampling rate)
    Fs = 1/fs                               # 取樣週期(sampling period)/取樣間隔(sampling interval)
    t = np.arange(0, Tc, Fs)                # 時間向量 (共256個點)
    Sample_rate = len(t)                    # 取樣頻率256個點
    bw = 500e+6                             # 發射訊號帶寬
    k = bw/Tc                               # 斜率
    P = 100                                 # chirp 的數目
    M = 4                                   # 接收天線數目
    N = 3                                   # 發射天線數目
    lamda = c/fc                            # 訊號波長
    # distance_temp = ((0:Sample_rate-1)*fs*c)/(2*k*Sample_rate);
    #----------------------------------------------------------------------------------------------------#
    point_cloud = []
    lambda_c = c / fc
    d = lambda_c / 2
    virtual_ant_num = N * M

    for det in detection_list:
        rbin = det["range_bin"]
        dbin = det["doppler_bin"]

        # 提取該點對應的虛擬天線資料：取出所有天線的複數值，shape=(M*N,)
        antenna_symbols = adc_data_cube[rbin, dbin, :]  # shape = (12,)

        # 重排成虛擬陣列：這裡假設為 2D 平面 (azimuth x elevation)，簡單重排為 3x4 (N x M)
        antenna_matrix = antenna_symbols.reshape(N, M)  # shape = (3, 4)

        # 做 2D FFT：對 azimuth (axis=1) & elevation (axis=0)
        fft_2d = np.fft.fft2(antenna_matrix, s=(N, M))
        fft_2d_shifted = np.fft.fftshift(np.abs(fft_2d))

        # 找最大峰值對應的位置（角度 bin）
        max_idx = np.unravel_index(np.argmax(fft_2d_shifted), fft_2d_shifted.shape)
        ele_bin, azi_bin = max_idx

        # 簡化估算角度（非精確）：從 FFT bin 轉為角度估計 (for illustration)
        azimuth_angle = np.arcsin((azi_bin - M//2) / (M/2)) * 180 / np.pi
        elevation_angle = np.arcsin((ele_bin - N//2) / (N/2)) * 180 / np.pi

        # 將 range 轉換為直角座標
        rng = (rbin * c) / (2 * k * Sample_rate / fs)
        x = rng * np.cos(np.deg2rad(elevation_angle)) * np.sin(np.deg2rad(azimuth_angle))
        y = rng * np.cos(np.deg2rad(elevation_angle)) * np.cos(np.deg2rad(azimuth_angle))
        z = rng * np.sin(np.deg2rad(elevation_angle))

        point_cloud.append({
            "range": rng,
            "azimuth_bin": azi_bin,
            "elevation_bin": ele_bin,
            "azimuth_angle_deg": azimuth_angle,
            "elevation_angle_deg": elevation_angle,
            "x": x,
            "y": y,
            "z": z,
            "magnitude": det["magnitude"]
        })
        
    # 轉換為 numpy 陣列方便處理
    xs = [pt["x"] for pt in point_cloud]
    ys = [pt["y"] for pt in point_cloud]
    zs = [pt["z"] for pt in point_cloud]
    mags = [pt["magnitude"] for pt in point_cloud]

    # 畫 3D 散點圖
    fig = plt.figure(figsize=(10, 8))
    ax = fig.add_subplot(111, projection='3d')
    sc = ax.scatter(xs, ys, zs, c=mags, cmap='viridis', s=50)

    ax.set_xlabel('X (m)')
    ax.set_ylabel('Y (m)')
    ax.set_zlabel('Z (m)')
    ax.set_title('3D Point Cloud from AoA2D DPU')
    ax.view_init(elev = 90, azim = 0)
    fig.colorbar(sc, label='Magnitude')
    plt.tight_layout()
    plt.show()

    return point_cloud

# 執行 AoA2D FFT 並轉為點雲
# point_cloud_list = aoa2d_dpu(adc_data_cube, range_doppler_detection_list)

# 顯示前幾個點雲結果
# point_cloud_list[:5]
```

```python
def aoa1d_fft_peak_estimation(adc_data_cube, detection_list, M=4, fft_points=256, Q=3):
    """
    1D AoA FFT over antenna array per CFAR detection point.
    Each detection uses IF_data[rbin, dbin, :] → [antennas]
    Perform 1D FFT and find Q strongest peaks in angle spectrum.

    Returns:
        point_cloud list with angle + (x, y) in 2D
    """
    c = 3e8
    fc = 77e9
    bw = 500e6
    fs = 25.6e6
    Tc = 10e-6
    M = 4
    N = 3
    k = bw / Tc
    Sample_rate = 256
    lamda = c / fc
    d = lamda / 2

    point_cloud = []

    for det in detection_list:
        rbin = det["range_bin"]
        dbin = det["doppler_bin"]

        # 取出該點對應天線的資料 (shape: [M,])
        
        antenna_symbols = adc_data_cube[rbin, dbin, :]  # shape = (12,)

        # 重排成虛擬陣列：這裡假設為 2D 平面 (azimuth x elevation)，簡單重排為 3x4 (N x M)
        antenna_matrix = antenna_symbols.reshape(N, M)  # shape = (3, 4)
        
        # antenna_vector = adc_data_cube[rbin, dbin, :M]  # 前 M 個接收天線
        

        # 每根天線做 FFT（對方向）
        # Y = np.tile(antenna_vector, (100, 1)).T  # 模擬重複 chirps 平均，這裡視情況調整
        Y = antenna_matrix[:, 0]
        spec = np.abs(np.fft.fftshift(np.fft.fft(Y, n=fft_points, axis=0), axes=0))
        # spec = np.mean(X_fft, axis=1)

        # 角度軸
        angle_axis = np.arcsin(np.linspace(-1, 1, fft_points)) * 180 / np.pi

        # 找 Q 個 peak
        from scipy.signal import find_peaks
        peaks, props = find_peaks(spec)
        top_peaks = sorted(peaks, key=lambda x: spec[x], reverse=True)[:Q]
        for pk in top_peaks:
            angle = angle_axis[pk]
            rng = (rbin * c) / (2 * k * Sample_rate / fs)
            x = rng * np.sin(np.deg2rad(angle))
            y = rng * np.cos(np.deg2rad(angle))
            point_cloud.append({
                "range": rng,
                "azimuth_angle_deg": angle,
                "x": x,
                "y": y,
                "z": 0,
                "magnitude": det["magnitude"]
            })

    # 畫角度光譜圖（使用最後一筆）
    plt.figure(figsize=(10, 4))
    plt.plot(angle_axis, spec, 'r-')
    plt.title("1D AoA Angle Spectrum")
    plt.xlabel("Angle (deg)")
    plt.ylabel("Power")
    plt.grid(True)
    plt.xlim([-90, 90])
    plt.xticks(np.arange(-90, 91, 10))
    plt.tight_layout()
    plt.show()

    return point_cloud
```
