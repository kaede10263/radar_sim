```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import LineString, Polygon, Point

# ==== 牆面開關 ====
enable_top_wall = False
enable_bottom_wall = False
enable_left_wall = False
enable_right_wall = False

# ==== 雷達參數 ====
radar = np.array([0.0, 0.0])
r_max = 200
angle_max = 60  
angle_min = -60
angle_scale = 5  

# ==== 目標物 ====
targets = []

circle_center = np.array([30, 6.5])
circle_radius = 2.0
circle_poly = Point(circle_center).buffer(circle_radius, resolution=32)
targets.append(circle_poly)

rect = Polygon([(20, -10), (26, -10), (26, -5), (20, -5)])
targets.append(rect)

triangle = Polygon([(40, 15), (43, 20), (37, 20)])
targets.append(triangle)

# ==== 牆定義 ====
walls = []
x_wall = 100
y_wall = 30

if enable_bottom_wall:
    walls.append(LineString([(-x_wall, -y_wall), (x_wall, -y_wall)]))
if enable_top_wall:
    walls.append(LineString([(-x_wall, y_wall), (x_wall, y_wall)]))
if enable_left_wall:
    walls.append(LineString([(-x_wall, -y_wall), (-x_wall, y_wall)]))
if enable_right_wall:
    walls.append(LineString([(x_wall, -y_wall), (x_wall, y_wall)]))

# ==== 畫圖初始化 ====
fig, ax = plt.subplots(figsize=(10, 6))
ax.set_aspect('equal')
plt.grid(True)
plt.title("Radar Reflection Simulation with Walls")
plt.xlabel("X")
plt.ylabel("Y")

for poly in targets:
    x, y = poly.exterior.xy
    ax.plot(x, y, 'k-')

for wall in walls:
    x, y = wall.xy
    ax.plot(x, y, 'b--', linewidth=1, label='Wall')

ax.plot(radar[0], radar[1], 'bo', label='Radar')

theta_list = {}  # 儲存紅線回波方向（弧度）
for single_deg in range(angle_min, angle_max, angle_scale):
    print(f"degree : {single_deg}")
    # ==== 發射光線 ====
    theta = np.deg2rad(single_deg)
    dir_vec = np.array([np.cos(theta), np.sin(theta)])
    current_point = radar.copy()
    current_dir = dir_vec.copy()
    total_length = 0
    reflection_points = []
    reflection_count = 0
    first_hit_red_line_drawn = False
    

    while total_length < r_max:
        ray_end = current_point + current_dir * (r_max - total_length)
        ray = LineString([current_point, ray_end])

        nearest_hit = None
        nearest_dist = np.inf
        nearest_poly = None

        # ==== 搜尋目標 ====
        for poly in targets:
            inter = ray.intersection(poly.boundary)
            print(f"inter.geom_type {inter.geom_type}")
            if inter.is_empty:
                continue
            if inter.geom_type == 'MultiPoint':
                for p in inter.geoms:
                    pt = np.array(p.coords[0])
                    dist = np.linalg.norm(pt - current_point)
                    if dist < nearest_dist and dist > 1e-6:
                        nearest_hit = pt
                        nearest_dist = dist
                        nearest_poly = poly
            elif inter.geom_type == 'Point':
                pt = np.array(inter.coords[0])
                dist = np.linalg.norm(pt - current_point)
                if dist < nearest_dist and dist > 1e-6:
                    nearest_hit = pt
                    nearest_dist = dist
                    nearest_poly = poly

        # ==== 搜尋牆 ====
        for wall in walls:
            inter = ray.intersection(wall)
            if inter.is_empty:
                continue
            if inter.geom_type == 'Point':
                pt = np.array(inter.coords[0])
                dist = np.linalg.norm(pt - current_point)
                if dist < nearest_dist and dist > 1e-6:
                    nearest_hit = pt
                    nearest_dist = dist
                    nearest_poly = 'wall'

        # ==== 沒碰到任何東西 ====
        if nearest_hit is None:
            final_point = current_point + current_dir * (r_max - total_length)
            ax.plot([current_point[0], final_point[0]], [current_point[1], final_point[1]], 'gray', linewidth=1)
            break

        segment_length = np.linalg.norm(nearest_hit - current_point)
        if total_length + segment_length > r_max:
            final_point = current_point + current_dir * (r_max - total_length)
            ax.plot([current_point[0], final_point[0]], [current_point[1], final_point[1]], 'gray', linewidth=1)
            break

        # ==== 畫灰色線段 ====
        ax.plot([current_point[0], nearest_hit[0]], [current_point[1], nearest_hit[1]], 'gray', linewidth=1)
        total_length += segment_length
        reflection_points.append(nearest_hit.tolist())
        reflection_count += 1



        # ==== 計算反射方向 ====
        if nearest_poly == 'wall':
            if np.abs(nearest_hit[1] - y_wall) < 1e-2 or np.abs(nearest_hit[1] + y_wall) < 1e-2:
                reflect = np.array([current_dir[0], -current_dir[1]])
            else:
                reflect = np.array([-current_dir[0], current_dir[1]])
        else:
            for seg_start, seg_end in zip(list(nearest_poly.exterior.coords[:-1]), list(nearest_poly.exterior.coords[1:])):
                edge = LineString([seg_start, seg_end])
                if edge.distance(Point(nearest_hit)) < 1e-3:
                    break
            edge_vec = np.array(seg_end) - np.array(seg_start)
            edge_vec /= np.linalg.norm(edge_vec)
            normal_vec = np.array([-edge_vec[1], edge_vec[0]])
            to_radar = radar - nearest_hit
            if np.dot(normal_vec, to_radar) < 0:
                normal_vec *= -1
            incident = (nearest_hit - current_point)
            incident /= np.linalg.norm(incident)
            reflect = incident - 2 * np.dot(incident, normal_vec) * normal_vec

            # ==== 第一個反射點畫紅線 ====
            if not first_hit_red_line_drawn:
                ax.plot([nearest_hit[0], radar[0]], [nearest_hit[1], radar[1]], 'red', linewidth=1.5)
                theta_list[nearest_poly].append({
                    np.arctan2(nearest_hit[1] - radar[1], nearest_hit[0] - radar[0])
                })
                first_hit_red_line_drawn = True

        # ==== 更新射線位置與方向 ====
        current_point = nearest_hit
        current_dir = reflect

        # ==== 若回到雷達方向，畫紅線並停止 ====
        to_radar = radar - current_point
        to_radar /= np.linalg.norm(to_radar)
        dot_return = np.dot(current_dir, to_radar)
        cos_threshold = np.cos(np.deg2rad(1))

        if dot_return > cos_threshold:
            print("larger line")
            ax.plot([current_point[0], radar[0]], [current_point[1], radar[1]], 'red', linewidth=1.5)
            theta_list[nearest_poly].append({
                np.arctan2(current_point[1] - radar[1], current_point[0] - radar[0])
            })
            # break

    # ==== 輸出結果 ====
    print(f"\n✅ Total reflections: {reflection_count}")
    for i, pt in enumerate(reflection_points):
        print(f"  {i+1}. Hit at {pt}")

#     print("\n📐 Radar return angles (theta in degrees):")
#     for th in theta_list:
#         print(f"  {np.rad2deg(th):.2f}°")


ax.legend()
plt.xlim(-5, 50)
plt.ylim(-30, 35)
plt.show()
```

```python
def simulate_radar_reflection_point(radar, r_max, angle_min, angle_max, angle_scale,
                               result_dict, walls):
    
    fig, ax = plt.subplots(figsize=(10, 6))
    ax.set_aspect('equal')
    plt.grid(True)
    plt.title("Radar Reflection Simulation with Walls")
    plt.xlabel("X")
    plt.ylabel("Y")
    
    structured_output = []
    theta_list = []

    for name, (poly, _, _) in result_dict.items():
        x, y = poly.exterior.xy
        ax.plot(x, y, 'k-')

    for wall in walls:
        x, y = wall.xy
        ax.plot(x, y, 'b--', linewidth=1, label='Wall')
    ax.plot(radar[0], radar[1], 'bo', label='Radar')

    special_case = 0

    for single_deg in range(angle_min, angle_max, angle_scale):
        # print(f"single_deg = {single_deg}")
        theta = np.deg2rad(single_deg)
        dir_vec = np.array([np.cos(theta), np.sin(theta)])
        current_point = radar.copy()
        current_dir = dir_vec.copy()
        total_length = 0

        while total_length < r_max:
            ray_end = current_point + current_dir * (r_max - total_length)
            ray = LineString([current_point, ray_end])

            nearest_hit = None
            nearest_dist = np.inf
            nearest_poly = None

            for name, (poly, _, _) in result_dict.items():
                inter = ray.intersection(poly.boundary)                
                if inter.is_empty or inter.geom_type in ['LineString', 'LinearRing']:
                    continue
                if inter.geom_type == 'MultiPoint':
                    for p in inter.geoms:                        
                        pt = np.array(p.coords[0])
                        dist = np.linalg.norm(pt - current_point)
                        if dist < nearest_dist and dist > 1e-6:
                            nearest_hit = pt
                            nearest_dist = dist
                            nearest_poly = poly
                            # print("in MultiPoint")
                            # print(f"{nearest_hit}, {nearest_dist}")
                                  
                elif inter.geom_type == 'Point':              
                    pt = np.array(inter.coords[0])
                    dist = np.linalg.norm(pt - current_point)
                    if dist < nearest_dist and dist > 1e-6:
                        nearest_hit = pt
                        nearest_dist = dist
                        nearest_poly = poly
                        # print("in Point")
                        # print(f"{nearest_hit}, {nearest_dist}")
                        
            for wall in walls:
                inter = ray.intersection(wall)
                if inter.is_empty:
                    continue
                if inter.geom_type == 'Point':
                    pt = np.array(inter.coords[0])
                    dist = np.linalg.norm(pt - current_point)
                    if dist < nearest_dist and dist > 1e-6:
                        nearest_hit = pt
                        nearest_dist = dist
                        nearest_poly = 'wall'
                        # print("in wall")
                        # print(f"{nearest_hit}, {nearest_dist}")

            if nearest_hit is None:
                final_point = current_point + current_dir * (r_max - total_length)
                ax.plot([current_point[0], final_point[0]], [current_point[1], final_point[1]], 'gray', linewidth=1)
                # print(f"line 83 draw gray and break (from {current_point[0]:.2f}, {current_point[1]:.2f} to {final_point[0]:.2f}, {final_point[1]:.2f})")
                break

            segment_length = np.linalg.norm(nearest_hit - current_point)
            if total_length + segment_length > r_max:
                final_point = current_point + current_dir * (r_max - total_length)
                ax.plot([current_point[0], final_point[0]], [current_point[1], final_point[1]], 'gray', linewidth=1)
                # print(f"line 89 draw gray and break (from {current_point[0]:.2f}, {current_point[1]:.2f} to {final_point[0]:.2f}, {final_point[1]:.2f})")
                break

            ax.plot([current_point[0], nearest_hit[0]], [current_point[1], nearest_hit[1]], 'gray', linewidth=1)
            total_length += segment_length
            # print(f"line 96 draw gray (from {current_point[0]:.2f}, {current_point[1]:.2f} to {nearest_hit[0]:.2f}, {nearest_hit[1]:.2f})")

            if nearest_poly == 'wall':
                if np.abs(nearest_hit[1] - y_wall) < 1e-2 or np.abs(nearest_hit[1] + y_wall) < 1e-2:
                    reflect = np.array([current_dir[0], -current_dir[1]])
                else:
                    reflect = np.array([-current_dir[0], current_dir[1]])
            else:
                for seg_start, seg_end in zip(list(nearest_poly.exterior.coords[:-1]), list(nearest_poly.exterior.coords[1:])):
                    edge = LineString([seg_start, seg_end])
                    if edge.distance(Point(nearest_hit)) < 1e-3:
                        break
                edge_vec = np.array(seg_end) - np.array(seg_start)
                edge_vec /= np.linalg.norm(edge_vec)
                normal_vec = np.array([-edge_vec[1], edge_vec[0]])
                to_radar = radar - nearest_hit
                if np.dot(normal_vec, to_radar) < 0:
                    normal_vec *= -1
                incident = (nearest_hit - current_point)
                incident /= np.linalg.norm(incident)
                reflect = incident - 2 * np.dot(incident, normal_vec) * normal_vec


            current_point = nearest_hit
            current_dir = reflect

            to_radar = radar - current_point
            to_radar /= np.linalg.norm(to_radar)
            dot_return = np.dot(current_dir, to_radar)
            cos_threshold = np.cos(np.deg2rad(1))

            special_case = (89 < abs(single_deg) <= 90 or 179 < abs(single_deg) <= 180)
            if special_case:
                break

            if (dot_return > cos_threshold and not special_case):
                ax.plot([current_point[0], radar[0]], [current_point[1], radar[1]], 'red', linewidth=1.5)
                name = [k for k, (v, _ , _) in result_dict.items() if v == nearest_poly]
                theta = np.round(np.rad2deg(np.arctan2(current_point[1] - radar[1], current_point[0] - radar[0])), 2)
                result_dict[name[0]][2].append(theta)
                if total_length < result_dict[name[0]][1][0]:
                    result_dict[name[0]][1][0] = np.round(total_length, 2)
                    result_dict[name[0]][1][1] = theta                    
                # print(f"line 133 draw red and break (from {current_point[0]:.2f}, {current_point[1]:.2f} to {radar[0]:.2f}, {radar[1]:.2f})")
                break
        # print()
        
    for target in targets:
        if isinstance(target, Point):
            continue 

        if not isinstance(target, Polygon):
            continue
        coords = list(target.exterior.coords)

        for i in range(len(coords) - 1):  # skip closing point
            p = np.array(coords[i])
            ray = LineString([radar, p])
            blocked = False

            # === 檢查被其他目標擋住 ===
            for poly in targets:
                if poly is target or isinstance(poly, Point):
                    continue
                inter = ray.intersection(poly)
                if not inter.is_empty and not inter.touches(Point(p)):
                    blocked = True
                    break

            # === 檢查被牆擋住 ===
            for wall in walls:
                if ray.crosses(wall):
                    blocked = True
                    break

            # === 自己遮擋自己 ===
            # print(f'{ray.crosses(target)}, {target.touches(Point(p))}')
            # if ray.crosses(target) and not target.touches(Point(p)):
            if ray.crosses(target) and target.touches(Point(p)):
                # print("ray crosses self")
                blocked = True

            if blocked:
                continue

            # === 法向量（順時針外法向量） ===
            p_prev = np.array(coords[i - 1])
            p_next = np.array(coords[(i + 1) % len(coords)])
            edge1 = p - p_prev
            edge2 = p_next - p
            n1 = np.array([edge1[1], -edge1[0]])
            n2 = np.array([edge2[1], -edge2[0]])
            normal = n1 + n2
            normal /= np.linalg.norm(normal)

            # === 雷達方向向量 ===
            ray_vec = radar - p
            ray_vec /= np.linalg.norm(ray_vec)

            dot = np.dot(normal, ray_vec)
            # print(normal, ray_vec, dot)
            if abs(dot) > 0.99:
                plt.plot([radar[0], p[0]], [radar[1], p[1]], 'orange')  # 回射
                dist = np.round(np.linalg.norm(p - radar), 2)
                name = [k for k, (v,_ , _) in result_dict.items() if v == target]
                theta = np.round(np.rad2deg(np.arctan2(p[1] - radar[1], p[0] - radar[0])), 2)
                result_dict[name[0]][2].append(theta)
                if dist < result_dict[name[0]][1][0]:
                    result_dict[name[0]][1][0] = dist
                    result_dict[name[0]][1][1] = theta
            else:
                plt.plot([radar[0], p[0]], [radar[1], p[1]], color='gray')
                pass


    

    plt.xlim(-5, 65)
    plt.ylim(-30, 35)
    plt.show()
    
    return result_dict
```
